```python
#| label: library_import
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
```

```python
#| label: random_data
#| fig-cap: "Fig 1- Plotting our randomly generated data"

# Reference- https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

# Let us now create a random dataset consisting of isotropic Gaussian blobs for clustering by using the make_blob() function.
# Visualization of our random dataset
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=1, random_state=20, cluster_std=.4, center_box=(20, 6))
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
#| label: dbscan

# We will fit the DBSCAN model with our input data X and make predictons on it
# Note noisy samples or outlier points are given the label -1.
pred_labels = DBSCAN(eps=0.3, min_samples=25).fit_predict(X)
```

```python
#| label: outliers_index

# Getting the index of the outlier points
outliers_index = []
for i in range(len(pred_labels)):
    if pred_labels[i] == -1:
        outliers_index.append(i)
        
# Displaying the index of outlier points
outliers_index     
```

```python
#| label: outliers_points

# Outlier points
outliers = X[outliers_index]
outliers
```

```python
#| label: outliers_visualization
#| fig-cap: "Fig 2- Plotting our randomly generated data and marking our outlier points as red"

# Let us now visualize our data and mark the outlier points as red
plt.scatter(X[:,0], X[:,1])
plt.scatter(outliers[:,0], outliers[:,1], color='r')
```
