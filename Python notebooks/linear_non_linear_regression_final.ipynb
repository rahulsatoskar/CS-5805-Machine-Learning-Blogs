```python
#| label: library_import

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
warnings.simplefilter("ignore")
```

```python
#| label: LR
#| fig-cap: "Fig 1- Linear regression visualization"

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = np.random.rand(150,1)

# Let us generate a function y which is a linear function of X with added Gaussian noise
y = 10 * X + 5 + np.random.randn(150, 1)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X, y)

# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
X1 = np.linspace(0, 1, 150).reshape(-1, 1)
y1_pred = LR.predict(X1)

plt.scatter(X, y)
# Let us plot a linear regression line for the points X in red
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

```python
#| label: LR1
#| fig-cap: "Fig 2- Linear regression visualization with more noise"

# Adding more noise to our data
# Linear regression visualization with more noise in our data

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = np.random.rand(150,1)

# Let us generate a function y which is a linear function of X with added Gaussian noise
y = 10 * X + 5 + 5 * np.random.randn(150, 1)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X, y)

X1 = np.linspace(0, 1, 150).reshape(-1, 1)
y1_pred = LR.predict(X1)

plt.scatter(X, y)
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

```python
#| label: LR2
#| fig-cap: "Fig 3- Linear regression visualization with less noise"

# Adding less noise to our data
# Linear regression visualization with less noise in our data

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = np.random.rand(150,1)

# Let us generate a function y which is a linear function of X with added Gaussian noise
y = 10 * X + 5 + 0.2 * np.random.randn(150, 1)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X, y)

X1 = np.linspace(0, 1, 150).reshape(-1, 1)
y1_pred = LR.predict(X1)

plt.scatter(X, y)
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

```python
#| label: LR3
#| fig-cap: "Fig 4- Linear regression visualization with quadratic data"

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = 4 * np.random.rand(80,1) - 2

# Let us generate a function y which is a quadratic function of X with added Gaussian noise
y = 2 * X + 5 + X**2 + np.random.randn(80, 1)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X, y)

# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
X1 = np.linspace(-2, 2, 80).reshape(-1, 1)
y1_pred = LR.predict(X1)

plt.scatter(X, y)
# Let us plot a linear regression line for the points X in red
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

Observations- We can see a quadratic curve and the best fit of line determined by linear regression model. As we can see a line does not do a good job in fitting a quadratic curve amd hence a linear model like linear regression is not good for fitting such a quadratic data. 

This is why we want to have polynomial regression to fit such a data. For example quadratic regression would be able to fit such a quadratic data.

For such a non linear data like the quadratic data as shown above we use polynomial features where we take the features which we have and we create polynomial features based on those features. 

```python
#| label: LR4
#| fig-cap: "Fig 5- Linear regression visualization with quadratic data and quadratic features used to train linear regression model"

from sklearn.preprocessing import PolynomialFeatures
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = 4 * np.random.rand(80,1) - 2

# Let us generate a function y which is a quadratic function of X with added Gaussian noise
y = 2 * X + 5 + X**2 + np.random.randn(80, 1)

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X_poly, y)

# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
X1 = np.linspace(-2, 2, 80).reshape(-1, 1)
X1_poly = poly.fit_transform(X1)

y1_pred = LR.predict(X1_poly)

plt.scatter(X, y)
# Let us plot a linear regression line for the points X in red
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

Now we can see we have a quadratic function to fit these points which does a better job than linear regression.

```python
#| label: LR5
#| fig-cap: "Fig 6- Linear regression visualization with a complex data and degree 4 features used to train linear regression model"


from sklearn.preprocessing import PolynomialFeatures
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
X = 4 * np.random.rand(80,1) - 2

# Let us generate a function y which has degree 1, quadratic, cubic and degree 4 terms of X with added Gaussian noise
y = 2 * X + 5 + X**2 + 10 * X**3 + 2 * X**4 + 5 * np.random.randn(80, 1)

poly = PolynomialFeatures(degree=4, include_bias=False)
X_poly = poly.fit_transform(X)

# Fitting a linear regression model to the generated dataset
LR = LinearRegression()
LR.fit(X_poly, y)

# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
X1 = np.linspace(-2, 2, 80).reshape(-1, 1)
X1_poly = poly.fit_transform(X1)

y1_pred = LR.predict(X1_poly)

plt.scatter(X, y)
# Let us plot a linear regression line for the points X in red
plt.plot(X1, y1_pred, color = 'r')
plt.show()
```

Observations- We can observe after generating degree 4 polynomial features we have fit our complex function relativey well.
